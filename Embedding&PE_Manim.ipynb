```python
%%manim -qm -v WARNING TokenToIDs

class TokenToIDs(Scene):
    def construct(self):
        # Create the "Original Sentence:" text
        original_sentence_text = Text("Original Sentence:", font_size=15)

        # Create a list of texts to put inside the rectangles
        texts = ["Your", "cat", "is", "a", "lovely", "cat"]

        # Create a list to hold the rectangles and arrows
        rectangles_and_arrows = []

        # Define the size of each rectangle
        rect_width = 1.2
        rect_height = 0.8

        # Create the rectangles with text inside and add them to the list
        for i, text in enumerate(texts):
            rect = Rectangle(width=rect_width, height=rect_height, color=BLUE)
            text_mob = Text(text).scale(0.5)
            # Center the text within the rectangle
            text_mob.move_to(rect.get_center())
            # Create the arrow
            arrow = Arrow(start=UP, end=DOWN, buff=0.3).next_to(rect, DOWN)
            # Group the rectangle, text, and arrow together
            group = VGroup(rect, text_mob, arrow)
            rectangles_and_arrows.append(group)

        # Arrange the rectangles and arrows next to each other
        rectangles_and_arrows_group = VGroup(*rectangles_and_arrows).arrange(RIGHT, buff=0.5)

        # Position the "Original Sentence:" text to the left of the rectangles
        original_sentence_text.next_to(rectangles_and_arrows_group, LEFT, buff=0.5)
        original_sentence_text.shift(UP *1)
        # Create a group containing both the text and the rectangles with arrows
        full_group = VGroup(original_sentence_text, rectangles_and_arrows_group)

        # Move the whole group slightly to the left
        full_group.shift(RIGHT * 1)

        # Create the "Input IDs" text
        input_ids_text = Text("Input IDs:", font_size=12)
        position_in_vocab_text = Text("(Position in the vocabulary)", font_size=12)

        # Group the "Input IDs" text and the position text together
        input_ids_full_text = VGroup(input_ids_text, position_in_vocab_text).arrange(DOWN, buff=0.1)

        # Create a list to hold the new rectangles
        input_id_rectangles = []

        # Create the new rectangles and add them to the list
        for i in range(6):
            embeddings = [105, 6587, 5475, 3578, 65, 6587]
            rect = Rectangle(width=rect_width, height=rect_height, color=GREEN)
            text_mob = Text(f"{embeddings[i]}").scale(0.5)
            # Center the text within the rectangle
            text_mob.move_to(rect.get_center())
            # Group the rectangle and text together
            group = VGroup(rect, text_mob)
            input_id_rectangles.append(group)

        # Arrange the new rectangles next to each other
        input_id_rectangles_group = VGroup(*input_id_rectangles).arrange(RIGHT, buff=0.5)

        # Position the new rectangles below the arrows
        input_id_rectangles_group.next_to(rectangles_and_arrows_group, DOWN, buff=0.5)

        # Position the "Input IDs" text to the left of the new rectangles
        input_ids_full_text.next_to(input_id_rectangles_group, LEFT, buff=0.5)

        # Create a group containing both the text and the new rectangles
        input_ids_group = VGroup(input_ids_full_text, input_id_rectangles_group)

        # Add the full group and input IDs group to the scene
        self.add(full_group, input_ids_group)
```

```python
%%manim -qm -v WARNING EmbeddingTables

class EmbeddingTables(Scene):
    def construct(self):
        # Define the data for six different tables
        data_list = [
            [["952"], ["627"], ["..."], ["112"], ["123"]],
            [["845"], ["293"], ["..."], ["450"], ["678"]],
            [["741"], ["159"], ["..."], ["753"], ["951"]],
            [["369"], ["258"], ["..."], ["147"], ["258"]],
            [["123"], ["456"], ["..."], ["789"], ["321"]],
            [["654"], ["987"], ["..."], ["321"], ["654"]]
        ]

        tables = []
        arrows = []

        for data in data_list:
            table = MathTable(data, include_outer_lines=True, v_buff=0.5)
            table.scale(0.5)  # Scale down the table for better fitting
            arrow = Arrow(start=UP, end=DOWN, buff=0.5).next_to(table, UP)
            tables.append(table)
            arrows.append(arrow)

        # Position the tables side by side
        tables_group = VGroup(*tables).arrange(RIGHT, buff=0.5).shift(RIGHT * 1)
        arrows_group = VGroup(*arrows).arrange(RIGHT, buff=1.3).next_to(tables_group, UP)

        # Create the text "Embedding" and "(Vector of size 512)" below it
        embedding_text = Text("Embedding", font_size=20).scale(0.75)
        vector_text = Text("(Vector of size 512)", font_size=20).scale(0.5)
        text_group = VGroup(embedding_text, vector_text).arrange(DOWN).next_to(tables_group, LEFT, buff=1)

        # Add the tables, arrows, and text to the scene
        self.add(tables_group, arrows_group, text_group)
```

```python
%%manim -qm -v WARNING TokenToIDsAndEmbeddingTables

class TokenToIDsAndEmbeddingTables(Scene):
    def construct(self):
        # "Orijinal Cmle:" metnini olutur
        original_sentence_text = Text("Original Sentence:", font_size=15)

        # Dikdrtgenlerin iine koyulacak metinlerin listesi
        texts = ["Your", "cat", "is", "a", "lovely", "cat"]

        # Dikdrtgenler ve oklar iin liste olutur
        rectangles_and_arrows = []

        # Her dikdrtgenin boyutunu belirle
        rect_width = 1.0
        rect_height = 0.5

        # Dikdrtgenleri metinlerle olutur ve listeye ekle
        for text in texts:
            rect = Rectangle(width=rect_width, height=rect_height, color=BLUE)
            text_mob = Text(text).scale(0.4)
            text_mob.move_to(rect.get_center())
            arrow = Arrow(start=UP, end=DOWN, buff=0.5).next_to(rect, DOWN)
            group = VGroup(rect, text_mob, arrow)
            rectangles_and_arrows.append(group)

        # Dikdrtgenleri ve oklar yan yana dzenle
        rectangles_and_arrows_group = VGroup(*rectangles_and_arrows).arrange(RIGHT, buff=0.3)

        # "Orijinal Cmle:" metnini dikdrtgenlerin soluna yerletir
        original_sentence_text.next_to(rectangles_and_arrows_group, LEFT, buff=0.3)
        original_sentence_text.shift(UP * 0.6)

        # "Orijinal Cmle:" metni ve dikdrtgenleri oklarla birlikte gruplandr
        full_group = VGroup(original_sentence_text, rectangles_and_arrows_group)
        full_group.shift(LEFT * 1)

        # "Girdi Kimlikleri" metnini olutur
        input_ids_text = Text("Input IDs:", font_size=12)
        position_in_vocab_text = Text("(Position in the vocabulary)", font_size=12)
        input_ids_full_text = VGroup(input_ids_text, position_in_vocab_text).arrange(DOWN, buff=0.1)

        # Yeni dikdrtgenler iin liste olutur
        input_id_rectangles = []

        # Embedding deerlerini tanmla
        embeddings = [105, 6587, 5475, 3578, 65, 6587]

        # Yeni dikdrtgenleri olutur ve listeye ekle
        for embedding in embeddings:
            rect = Rectangle(width=rect_width, height=rect_height, color=GREEN)
            text_mob = Text(f"{embedding}").scale(0.4)
            text_mob.move_to(rect.get_center())
            group = VGroup(rect, text_mob)
            input_id_rectangles.append(group)

        # Yeni dikdrtgenleri yan yana dzenle
        input_id_rectangles_group = VGroup(*input_id_rectangles).arrange(RIGHT, buff=0.3)
        input_id_rectangles_group.next_to(rectangles_and_arrows_group, DOWN, buff=0.5)
        input_ids_full_text.next_to(input_id_rectangles_group, LEFT, buff=0.3)

        # Yeni dikdrtgenleri ve metni gruplandr
        input_ids_group = VGroup(input_ids_full_text, input_id_rectangles_group)

        # Alt farkl tablo iin verileri tanmla
        data_list = [
            [["952"], ["627"], ["..."], ["112"], ["123"]],
            [["845"], ["293"], ["..."], ["450"], ["678"]],
            [["741"], ["159"], ["..."], ["753"], ["951"]],
            [["369"], ["258"], ["..."], ["147"], ["258"]],
            [["123"], ["456"], ["..."], ["789"], ["321"]],
            [["845"], ["293"], ["..."], ["450"], ["678"]]
        ]

        tables = []
        table_arrows = []

        for data in data_list:
            table = MathTable(data, include_outer_lines=True, v_buff=0.5)
            table.scale(0.5)
            tables.append(table)

        # Tablolar yan yana dzenle
        tables_group = VGroup(*tables).arrange(RIGHT, buff=0.3).shift(RIGHT * 1)

        # Tablolar iin oklar olutur
        for table in tables:
            arrow = Arrow(start=UP, end=DOWN, buff=0.5).next_to(table, UP)
            table_arrows.append(arrow)

        # Oklar gruplandr ve yerletir
        table_arrows_group = VGroup(*table_arrows).arrange(RIGHT, buff=1.05).next_to(tables_group, UP)
        # "Embedding" ve "(Vector of size 512)" metinlerini olutur
        embedding_text = Text("Embedding", font_size=15).scale(0.75)
        vector_text = Text("(Vector of size 512)", font_size=15).scale(0.75)
        text_group = VGroup(embedding_text, vector_text).arrange(DOWN).next_to(tables_group, LEFT, buff=0.5)

        # Tablolar ve oklar input_id_rectangles_group'un altna yerletir
        tables_group.next_to(input_id_rectangles_group, DOWN, buff=1.4)
        table_arrows_group.next_to(tables_group, UP)
        text_group.next_to(tables_group, LEFT, buff=1)

        # Tm bileenleri gruplandr
        all_components = VGroup(full_group, input_ids_group, tables_group, table_arrows_group, text_group)

        # Tm bileenleri sahnenin ortasna ta
        all_components.move_to(ORIGIN)

        # Tm bileenleri sahneye ekle
        self.add(all_components)
```

```python
%%manim -qm -v WARNING PositionalEncoding

class PositionalEncoding(Scene):
    def construct(self):
        formulas = [
            MathTex(r"\text{PE}_{(pos, 2i)} = \sin\left(\frac{pos}{10000^{2i\over d_{model}}}\right)"),
            MathTex(r"\text{PE}_{(pos, 2i+1)} = \cos\left(\frac{pos}{10000^{2i\over d_{model}}}\right)")
        ]

        formulas[0].next_to(formulas[1], UP, buff=1)
        formulas[1].next_to(formulas[0], DOWN, buff=1)

        for formula in formulas:
            self.add(formula)
```
